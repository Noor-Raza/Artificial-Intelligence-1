````python
# DO NOT MODIFY NOR ADD CODE TO THIS CELL

df = pd.read_csv('https://raw.githubusercontent.com/jnin/information-systems/refs/heads/main/data/superheroes_complete.csv')
````

````python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline


### 1. Feature Selection and Engineering

X = df.drop(columns=['Alignment',                                                                               # Dropping the target variable
                     'Name', 'Publisher',                                                                       # Dropping irrelevant variables 'Name' and 'Publisher', because they are unique identifiers not influencing target variable
                     'Skin color'])                                                                             # Dropping irrelevant variable 'Skin Color' because it has 588 missing values and thus is insignificant
y = df['Alignment']                                                                                             # Setting the target variable


### 2. Handling Missing Data

mean_imputer = SimpleImputer(strategy='mean')                                                                   # Creating mean_imputer for numerical features
mode_imputer = SimpleImputer(strategy='most_frequent')                                                          # Creating mode_imputer for categorical features


### 3. Encoding Categorical and Ordinal Features

categories = [['Weak', 'Below Average', 'Average', 'Above Average', 'Extremely Powerful'],                      # Define ranking of the 'Power Level' categories
              ['Low Intelligence', 'Average Intelligence', 'Smart', 'Genius', 'Super-Genius']]                  # Define ranking of the 'Intelligence Level' categories

one_hot_encoder = OneHotEncoder(sparse_output=False)                                                            # Creating one_hot_encoder function
ordinal_encoder = OrdinalEncoder(categories=categories)                                                         # Creating ordinal_encoder function


### 4. Standardizing Features

scaler = StandardScaler()                                                                                       # Creating scaler function


### 5. Building a Pipeline

categorical_columns = ['Gender', 'Eye color', 'Race', 'Hair color']                                             # Defining categorical columns
ordinal_columns = ['Power Level', 'Intelligence Level']                                                         # Defining ordinal columns
numerical_columns = [col for col in X.columns if col not in categorical_columns + ordinal_columns]              # Defining all other columns as numerical

inner_pipe_categorical = Pipeline(steps=                                                                        # Creating categorical pipeline
                                  [('mode_imputer', mode_imputer),                                              # Imputing categorical values with the most frequent
                                   ('one_hot_encoder', one_hot_encoder)])                                       # Encoding categorical values with one_hot_encoder
inner_pipe_ordinal = Pipeline(steps=                                                                            # Creating ordinal pipeline
                              [('mode_imputer', mode_imputer),                                                  # Imputing ordinal values with the most frequent
                               ('ordinal_encoder', ordinal_encoder)])                                           # Encoding ordinal values with ordinal_encoder
inner_pipe_numerical = Pipeline(steps=                                                                          # Creating numerical pipeline
                                [('mean_imputer', mean_imputer)])                                               # Imputing numerical values with the mean

pipe_transformer = ColumnTransformer(                                                                           # Combining all pipelines in the column transformer
    [('inner_pipe_categorical', inner_pipe_categorical, categorical_columns),                                   # Adding categorical pipeline
     ('inner_pipe_ordinal', inner_pipe_ordinal, ordinal_columns),                                               # Adding ordinal pipeline
     ('inner_pipe_numerical', inner_pipe_numerical, numerical_columns)],                                        # Adding numerical pipeline
     remainder='passthrough')                                                                                   # Adding all remaining columns without being transformed

pipe = Pipeline(                                                                                                # Creating final model pipeline
    [('preprocessing', pipe_transformer),                                                                       # Adding preprocessing transformer
     ('scaler', scaler)])                                                                                       # Standardizing the results

X_pipe = pipe.fit_transform(X)                                                                                  # Fitting the model to the dataframe

X_pipe                                                                                                          # Displaying the results

### 6. Documentation -> see comments
````
